```python
import pandas as pd
import sqlite3

# SQLite dosyasının yolu
sqlite_file = "v12.gts.sqlite3.db"  # SQLite dosya adınızı buraya yazın

# SQLite bağlantısı oluştur
connection = sqlite3.connect(sqlite_file)

# SQLite dosyasındaki tabloların isimlerini öğren
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", connection)
print("Tablolar:")
print(tables)

# Bir tabloyu pandas DataFrame'e yükle
table_name = "madde"  # Okumak istediğiniz tablo adını buraya yazın
df = pd.read_sql_query(f"SELECT * FROM {table_name};", connection)
df
```

```python
# read taki column of madde count each unique value
kelimeler = set(madde['madde'].unique().tolist())
kelimeler_lowercase = set([str(k).lower() for k in kelimeler if k is not None])
kelimeler_lowercase
```

```python
import re

split_kelimeler = set()

for k in kelimeler_lowercase:
    splits = k.split(' ')
    # remove numbers and special characters
    splits = [re.sub(r'[^a-zA-ZğışüöçĞİŞÜÖÇ]+', '', s) for s in splits]
    split_kelimeler.update(splits)

split_kelimeler
```

```python
with open('kelimeler.txt', 'w', encoding='utf-8') as f:
    # sort by length and then alphabetically
    ordered_split_kelimeler = sorted(split_kelimeler, key=lambda x: (len(x), x))
    for k in ordered_split_kelimeler:
        f.write(k + '\n')
```

```python
copy_kelimeler = split_kelimeler.copy()
```

```python
# remove laştırılmak at the end of the word
# first check if the word ends with laştırılmak
# if it does, remove it
cleaned_kelimeler = set()
suffix_to_remove = 'yicilik'
for k in copy_kelimeler:
  if k is None or len(k) < 2:
    continue
  if len(k) > len(suffix_to_remove) + 2 and k.endswith(suffix_to_remove):
    cleaned_kelimeler.add(k[:-len(suffix_to_remove)])
  else:
    cleaned_kelimeler.add(k)

copy_kelimeler = cleaned_kelimeler.copy()

with open('clean_kelimeler.txt', 'w', encoding='utf-8') as f:
    # sort by length and then alphabetically
    ordered_cleaned_kelimeler = sorted(cleaned_kelimeler, key=lambda x: (len(x), x), reverse=True)
    for k in ordered_cleaned_kelimeler:
        f.write(k + '\n')
        
len(cleaned_kelimeler)
```
