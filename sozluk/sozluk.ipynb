```python
import pandas as pd
import sqlite3

# SQLite dosyasının yolu
sqlite_file = "v12.gts.sqlite3.db"  # SQLite dosya adınızı buraya yazın

# SQLite bağlantısı oluştur
connection = sqlite3.connect(sqlite_file)

# SQLite dosyasındaki tabloların isimlerini öğren
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", connection)
print("Tablolar:")
print(tables)

# Bir tabloyu pandas DataFrame'e yükle
table_name = "madde"  # Okumak istediğiniz tablo adını buraya yazın
df = pd.read_sql_query(f"SELECT * FROM {table_name};", connection)
df
```

```python
for table_name in tables["name"]:
    df = pd.read_sql_query(f"SELECT * FROM {table_name};", connection)
    print(f"\n{table_name} tablosu:", df.shape)
    # save to csv
    file_name = f"{table_name}.csv"
    df.to_csv(file_name, index=False, encoding="utf-8")
    print(df.head(2))
```
